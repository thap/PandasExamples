```python
editions = pd.read_csv('Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - EDITIONS.tsv',
                       sep='\t')
editions
```

```python
ioc_codes = pd.read_csv('Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - IOC COUNTRY CODES.csv',)
ioc_codes
```

```python
medals = pd.read_csv('Summer Olympic medals/Summer Olympic medalists 1896 to 2008 - ALL MEDALISTS.tsv', 
                     header=4, sep='\t')
medals
```

```python
medal_counts = medals.pivot_table(aggfunc='count', index='Edition', values='Athlete', columns='NOC')
medal_counts
```

```python
totals = editions.set_index('Edition')['Grand Total']
totals
```

```python
fractions = medal_counts.divide(totals, axis=0)
fractions
```

```python
mean_fractions = fractions.expanding().mean()
mean_fractions
```

```python
fractions_change = mean_fractions.pct_change() * 100
fractions_change = fractions_change.reset_index()
fractions_change
```

```python
editions, ioc_codes
```

```python
hosts = pd.merge(editions, ioc_codes, how='left')
hosts
```

```python
hosts.set_index('Edition', inplace=True)
hosts.loc[hosts.NOC.isnull()]
```

```python
hosts.loc[1972, 'NOC'] = 'FRG'
hosts.loc[1980, 'NOC'] = 'URS'
hosts.loc[1988, 'NOC'] = 'KOR'
hosts.loc[1972, 'ISO code'] = 'DE'
hosts.loc[1980, 'ISO code'] = 'SU'
hosts.loc[1988, 'ISO code'] = 'KR'
hosts = hosts.reset_index()
hosts
```

```python
ioc_codes[ioc_codes['NOC'].isin(['FRG', 'USR', 'KOR', 'GER', 'RUS'])]
```

```python
reshaped = pd.melt(fractions_change, id_vars='Edition', value_name='Change')
reshaped
```

```python
reshaped.shape, fractions_change.shape
```

```python
reshaped[reshaped['NOC'] == 'CHN'].dropna()
```

```python
merged = pd.merge(reshaped, hosts)
merged
```

```python
influence = merged.set_index('Edition').sort_index()
influence
```

```python
change = influence['Change']
ax = change.plot(kind='bar')
ax.set_ylabel('% Change of Host Country Medal Count')
ax.set_title('Is there a Host Country Advantage?')
ax.set_xticklabels(editions['City'])
plt.show()
```
