```python
def GrowUp(x):
    avg_weight = sum(x[x['size'] == 'S'].weight * 1.5)
    avg_weight += sum(x[x['size'] == 'M'].weight * 1.25)
    avg_weight += sum(x[x['size'] == 'L'].weight)
    avg_weight /= len(x)
    return pd.Series(['L', avg_weight, True], index=['size', 'weight', 'adult'])
```

```python
gb.apply(GrowUp)
```

```python
import functools
S = pd.Series([i / 100.0 for i in range(1, 11)])
def cum_ret(x, y):
    return x * (1 + y)
def red(x):
    return functools.reduce(cum_ret, x, 1.0)
```

```python
S.expanding().apply(red, raw=True)
```

```python
grouped = titanic.groupby(['sex', 'pclass'])
def impute_median(series):
    return series.fillna(series.median())
titanic['age'] = grouped['age'].transform(impute_median)
titanic.info()
```

```python
def disparity(gr):
    s = gr['gdp'].max() - gr['gdp'].min()
    z = (gr['gdp'] - gr['gdp'].mean())/gr['gdp'].std()
    return pd.DataFrame({'z(gdp)':z , 'regional spread(gdp)':s})
```

```python
gap
```

```python
regional_disparity = gap.groupby('region').apply(disparity)
```

```python
gap = pd.concat([gap, regional_disparity], axis='columns')
```

```python
gap[gap['Country'].isin(['China'])]
```

```python
gap[gap['Country'].isin(['United States'])]
```

```python
def c_deck_survival(gr):
    c_passengers = gr['cabin'].str.startswith('C').fillna(False)
    return gr.loc[c_passengers, 'survived'].mean()
```

```python
titanic.groupby('sex').apply(c_deck_survival)
```
